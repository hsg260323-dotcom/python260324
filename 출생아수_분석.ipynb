# 대한민국 출생아수 및 합계출산율 데이터 분석

이 노트북에서는 대한민국의 출생아수와 합계출산율 데이터를 pandas로 분석하고 시각화합니다.

## 1. 라이브러리 임포트

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Microsoft YaHei'  # Windows의 기본 한글 폰트
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호 처리
```

## 2. Excel 파일 읽기

```python
# Excel 파일 읽기
file_path = r'C:\work\출생아수__합계출산율__자연증가_등_20260327151751.xlsx'
df = pd.read_excel(file_path)

# 원본 데이터 확인
print("원본 데이터 크기:", df.shape)
print("\n원본 데이터 처음 5행:")
print(df.head())
print("\n데이터 컬럼 정보:")
print(df.info())
```

## 3. 데이터 클랜징

```python
# 데이터 복사
df_clean = df.copy()

# 1. 결측값 확인 및 처리
print("결측값 확인:")
print(df_clean.isnull().sum())
print()

# 2. 중복값 확인 및 제거
print(f"중복행 개수: {df_clean.duplicated().sum()}")
df_clean = df_clean.drop_duplicates()
print()

# 3. 컬럼명 정리 (공백 제거, 소문자 변환)
df_clean.columns = df_clean.columns.str.strip()
print("정리된 컬럼명:")
print(df_clean.columns.tolist())
print()

# 4. 년도 컬럼 찾기 및 정렬
# 일반적인 년도 컬럼명으로 시작하는 컬럼 찾기
year_col = None
birth_col = None
tfr_col = None  # 합계출산율

for col in df_clean.columns:
    if '년도' in col or '년' in col:
        year_col = col
    if '출생' in col and '수' in col:
        birth_col = col
    if '합계출산율' in col or '출산율' in col:
        tfr_col = col

print(f"년도 컬럼: {year_col}")
print(f"출생아수 컬럼: {birth_col}")
print(f"합계출산율 컬럼: {tfr_col}")
print()

# 5. 데이터 타입 변환
if year_col:
    df_clean[year_col] = pd.to_numeric(df_clean[year_col], errors='coerce')
    
if birth_col:
    df_clean[birth_col] = pd.to_numeric(df_clean[birth_col], errors='coerce')
    
if tfr_col:
    df_clean[tfr_col] = pd.to_numeric(df_clean[tfr_col], errors='coerce')

print("변환 후 데이터 타입:")
print(df_clean.dtypes)
print()

# 6. 클랜징 후 결측값 확인
print("클랜징 후 결측값:")
print(df_clean.isnull().sum())
print()

print("정제된 데이터 처음 5행:")
print(df_clean.head(10))
```

## 4. 기초 통계 분석

```python
# 기초 통계 정보
print("=" * 60)
print("기초 통계 정보")
print("=" * 60)
print(df_clean.describe())
print()

# 결측값 제거 후 분석
df_analysis = df_clean.dropna()
print("=" * 60)
print(f"데이터 크기: {df_analysis.shape[0]}행 × {df_analysis.shape[1]}열")
print("=" * 60)
print()

if birth_col:
    print(f"출생아수 통계:")
    print(f"  - 최대값: {df_analysis[birth_col].max():,.0f}")
    print(f"  - 최소값: {df_analysis[birth_col].min():,.0f}")
    print(f"  - 평균값: {df_analysis[birth_col].mean():,.0f}")
    print(f"  - 표준편차: {df_analysis[birth_col].std():,.0f}")
    print()

if tfr_col:
    print(f"합계출산율 통계:")
    print(f"  - 최대값: {df_analysis[tfr_col].max():.3f}")
    print(f"  - 최소값: {df_analysis[tfr_col].min():.3f}")
    print(f"  - 평균값: {df_analysis[tfr_col].mean():.3f}")
    print(f"  - 표준편차: {df_analysis[tfr_col].std():.3f}")
```

## 5. 다각도 데이터 분석

```python
# 년도별 데이터 정렬
if year_col:
    df_analysis = df_analysis.sort_values(by=year_col)
    
    print("=" * 60)
    print("1. 년도별 출생아수 추이")
    print("=" * 60)
    
    if birth_col:
        year_birth = df_analysis[[year_col, birth_col]].copy()
        year_birth.columns = ['년도', '출생아수']
        print(year_birth.to_string(index=False))
        print()

# 년도별 변화율 분석
if year_col and birth_col:
    print("=" * 60)
    print("2. 년도별 출생아수 변화율 분석")
    print("=" * 60)
    
    df_growth = df_analysis[[year_col, birth_col]].copy()
    df_growth.columns = ['년도', '출생아수']
    df_growth['변화량'] = df_growth['출생아수'].diff()
    df_growth['변화율(%)'] = df_growth['출생아수'].pct_change() * 100
    
    print(df_growth.to_string(index=False))
    print()

# 출생아수와 합계출산율의 상관관계
if birth_col and tfr_col:
    print("=" * 60)
    print("3. 출생아수와 합계출산율의 상관관계")
    print("=" * 60)
    
    correlation = df_analysis[[birth_col, tfr_col]].corr()
    print(correlation)
    print(f"\n출생아수와 합계출산율의 상관계수: {df_analysis[birth_col].corr(df_analysis[tfr_col]):.4f}")
    print()

# 연도별 구간별 분석
if year_col and birth_col:
    print("=" * 60)
    print("4. 연도별 구간 분석")
    print("=" * 60)
    
    df_period = df_analysis[[year_col, birth_col]].copy()
    df_period.columns = ['년도', '출생아수']
    
    # 연도별 구간 분류
    df_period['구간'] = pd.cut(df_period['년도'], 
                               bins=[df_period['년도'].min()-1, 2000, 2010, 2020, df_period['년도'].max()],
                               labels=['~2000년', '2001-2010', '2011-2020', '2021년~'])
    
    print(df_period.groupby('구간')['출생아수'].agg(['count', 'mean', 'min', 'max']))
    print()
```

## 6. 년도별 출생아수 라인 그래프

```python
# 년도별 출생아수 라인 그래프
if year_col and birth_col:
    fig, ax = plt.subplots(figsize=(14, 7))
    
    plot_data = df_analysis[[year_col, birth_col]].sort_values(by=year_col)
    
    ax.plot(plot_data[year_col], plot_data[birth_col], 
            marker='o', linewidth=2.5, markersize=6, color='#1f77b4', label='출생아수')
    
    ax.set_xlabel('연도', fontsize=12, fontweight='bold')
    ax.set_ylabel('출생아수 (명)', fontsize=12, fontweight='bold')
    ax.set_title('대한민국 연도별 출생아수 추이', fontsize=14, fontweight='bold', pad=20)
    
    # 그리드 추가
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # y축에 천 단위 구분
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))
    
    # x축 레이블 회전
    plt.xticks(rotation=45)
    
    # 범례 추가
    ax.legend(loc='best', fontsize=11)
    
    plt.tight_layout()
    plt.show()
    
    print("라인 그래프 생성 완료!")
    print()
```

## 7. 추가 시각화 분석

```python
# 여러 개의 시각화를 함께 표시
if year_col and birth_col and tfr_col:
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    plot_data = df_analysis[[year_col, birth_col, tfr_col]].sort_values(by=year_col)
    
    # 1. 출생아수 라인 그래프
    axes[0, 0].plot(plot_data[year_col], plot_data[birth_col], 
                    marker='o', linewidth=2, markersize=5, color='#1f77b4')
    axes[0, 0].set_title('출생아수 추이', fontsize=11, fontweight='bold')
    axes[0, 0].set_xlabel('연도')
    axes[0, 0].set_ylabel('출생아수 (명)')
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))
    
    # 2. 합계출산율 라인 그래프
    axes[0, 1].plot(plot_data[year_col], plot_data[tfr_col], 
                    marker='s', linewidth=2, markersize=5, color='#ff7f0e')
    axes[0, 1].set_title('합계출산율 추이', fontsize=11, fontweight='bold')
    axes[0, 1].set_xlabel('연도')
    axes[0, 1].set_ylabel('합계출산율')
    axes[0, 1].grid(True, alpha=0.3)
    
    # 3. 출생아수 바 그래프
    axes[1, 0].bar(plot_data[year_col], plot_data[birth_col], color='#2ca02c', alpha=0.7, width=0.6)
    axes[1, 0].set_title('연도별 출생아수 (바 그래프)', fontsize=11, fontweight='bold')
    axes[1, 0].set_xlabel('연도')
    axes[1, 0].set_ylabel('출생아수 (명)')
    axes[1, 0].grid(True, alpha=0.3, axis='y')
    axes[1, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))
    
    # 4. 출생아수와 합계출산율 비교 (이중 축)
    ax1 = axes[1, 1]
    ax2 = ax1.twinx()
    
    line1 = ax1.plot(plot_data[year_col], plot_data[birth_col], 
                     marker='o', linewidth=2, markersize=5, color='#1f77b4', label='출생아수')
    line2 = ax2.plot(plot_data[year_col], plot_data[tfr_col], 
                     marker='s', linewidth=2, markersize=5, color='#ff7f0e', label='합계출산율')
    
    ax1.set_xlabel('연도')
    ax1.set_ylabel('출생아수 (명)', color='#1f77b4')
    ax2.set_ylabel('합계출산율', color='#ff7f0e')
    ax1.set_title('출생아수 vs 합계출산율', fontsize=11, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))
    
    # 범례
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper left', fontsize=9)
    
    # x축 레이블 회전
    for ax in axes.flat:
        plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    print("추가 시각화 완료!")
```
